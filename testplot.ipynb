```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.loadtxt("testresult.txt", dtype=np.int, delimiter=',')
print(results)
```

```python
f, axs = plt.subplots(1, 4, sharey=True)
colors = ('r', 'm', 'y', 'b')
for i in range(len(axs)):
    bars = axs[i].bar(np.arange(4), results[i], width=1, color='b',
                        align="center", tick_label=('Dress', 'Nude', 'School', 'Swimsuit'))
    for j in range(len(bars)):
        # Each bar has its own color.
        bars[j].set_color(colors[j])
        height = bars[j].get_height()
        center = bars[j].get_x() + bars[j].get_width() / 2.
        # Each bar has its percentage written above.
        axs[i].text(center, height, "{:.2f}%".format(results[i][j] / sum(results[i]) * 100),
                    ha='center', va='bottom')
f.set_figwidth(16)
axs[0].set_ylabel("Predictions")
plt.show()
```

```python
data = results[0]
bars = plt.barh(np.arange(len(data)), data, tick_label=('Dress', 'Nude', 'School', 'Swimsuit'), align="center", height=1)
for i in range(len(bars)):
    bars[i].set_color(colors[i])
plt.show()
```
